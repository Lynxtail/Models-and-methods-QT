## Цепи с непрерывным временем

### 1.

```python
L = np.array([[-4-7, 4, 7, 0],
             [6, -6-1-2, 1, 2],
             [8, 9, -8-9-4, 4],
             [0, 1, 3, -3-1]], dtype=np.float)
```

### 2.

```python
expm(L*10000)
```

Предельное распределение найдено

### 3.

```python
def exp_rand(lmbd, eps):
    return -1/lmbd * np.log(random.random()+eps)
```

```python
def get_state(i, P):
    p = random.random()
    last = 0
    ans = i
    for j in range(len(P)):
        if last <= p <= last+P[i,j]: 
            ans = j 
            break
        last += P[i,j]
    return ans
```

```python
def get_p(i,j):
    if i != j:
        return L[i,j]/(-L[i,i])
    else: return 0
```

```python
eps = 0.0001
T_max = 10000
P = [[],[],[],[]]
for i in range(len(P)):
    for j in range(len(P)):
        P[i].append(get_p(i,j))
P = np.array(P)
print(P)
cur_state = 0
q = np.abs(np.diagonal(L))
times = [exp_rand(q[cur_state], eps), 0, 0, 0]
cur_time = times[cur_state]

while (cur_time <= T_max):
    cur_state = get_state(cur_state, P)
    time_in_state = exp_rand(q[cur_state], eps)
    times[cur_state] += time_in_state
    cur_time += time_in_state
    #print(cur_time)
print("\nState's time: ", times, '\nSummary time: ', sum(times))
ans = [item/sum(times) for item in times]
print('\nStationary distribution: ', ans)
```
